```python
df_data.drop('id',axis=1, inplace =True) #droping column 'id' as default index is present
df_data.head()
```

```python
#updating column names as understandable
df_data.columns = ['age', 'blood_pressure', 'specific_grevity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
                   'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
                   'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
                   'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
                   'anemia', 'class']
```

```python
df_data.head()
```

```python
text_columns = ['packed_cell_volume','white_blood_cell_count','red_blood_cell_count'] #changing the dtypeback to float
for i in text_columns:
    print(f'{i}:{df_data[i].dtype}')
```

```python
def text_to_numeric(df_data, column):
    df_data[column] = pd.to_numeric(df_data[column], errors='coerce')

for column in text_columns:
    text_to_numeric(df_data, column)
    print(f'{column}:{df_data[column].dtype}')
```

```python
missing = df_data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
def mean_value_imputation(df_data, column): #Average of the values
    mean_value = df_data[column].mean()
    df_data[column].fillna(value=mean_value, inplace=True)

def mode_value_imputation(df_data, column): #Values with maximum number
    mode = df_data[column].mode()[0]
    df_data[column] = df_data[column].fillna(mode)
```

```python
num_cols =  [col for col in df_data.columns if df_data[col].dtype !='object'] #adding numerical columns in to a list using list comprehention
for col_name in num_cols:
    mean_value_imputation(df_data, col_name)   #replacing missing values(NA) with mean
```

```python
cat_cols =  [col for col in df_data.columns if df_data[col].dtype =='object'] #adding non-numerical(categorical) columns in to a list using list comprehention
for col_name in cat_cols:
    mode_value_imputation(df_data, col_name)   #replacing missing values(NA) with mode
```

```python
missing = df_data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
print(f"diabetes_mellitus : {df_data['diabetes_mellitus'].unique()}")
print(f"coronary_artery_disease : {df_data['coronary_artery_disease'].unique()}")
print(f"class : {df_data['class'].unique()}")

#for col in cat_cols:
#    unique_vals = df_data[col].unique()
#    print(f"{col} : {unique_vals}")
```

```python
df_data['diabetes_mellitus']=df_data['diabetes_mellitus'].replace(to_replace={' yes':'yes', '\tno':'no', '\tyes':'yes'})
df_data['coronary_artery_disease']=df_data['coronary_artery_disease'].replace(to_replace={' yes':'yes', '\tno':'no', '\tyes':'yes'})
df_data['class']=df_data['class'].replace(to_replace={'ckd\t':'ckd', 'notckd':'not ckd'})
```

```python
##Feature Encoding -- changing the categorical values to numeric
df_data['class'] = df_data['class'].map({'ckd':1, 'not ckd': 0})
df_data['red_blood_cells'] = df_data['red_blood_cells'].map({'normal':1, 'abnormal': 0})
df_data['pus_cell'] = df_data['pus_cell'].map({'normal':1, 'abnormal': 0})
df_data['pus_cell_clumps'] = df_data['pus_cell_clumps'].map({'present':1, 'notpresent': 0})
df_data['bacteria'] = df_data['bacteria'].map({'present':1, 'notpresent': 0})
df_data['hypertension'] = df_data['hypertension'].map({'yes':1, 'no': 0})
df_data['diabetes_mellitus'] = df_data['diabetes_mellitus'].map({'yes':1, 'no': 0})
df_data['coronary_artery_disease'] = df_data['coronary_artery_disease'].map({'yes':1, 'no': 0})
df_data['appetite'] = df_data['appetite'].map({'good':1, 'poor': 0})
df_data['peda_edema'] = df_data['peda_edema'].map({'yes':1, 'no': 0})
df_data['anemia'] = df_data['anemia'].map({'yes':1, 'no': 0})
```

```python
#building the model

from sklearn.model_selection import train_test_split
X = df_data.drop("class",axis=1)
y = df_data['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=25)

print(f"'X' shape: {X_train.shape}")
print(f"'Xtest' shape: {X_test.shape}")
```

```python
from sklearn.tree import DecisionTreeClassifier

dct = DecisionTreeClassifier()
dct.fit(X_train, y_train)
```

```python
y_pred_dct = dct.predict(X_test)
y_pred_dct
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC


models = []
models.append(('Naive Bayes', GaussianNB()))
models.append(('KNN', KNeighborsClassifier(n_neighbors=8)))  #Neighbours should be given while creating KNN
models.append(('RandomForestClassifier', RandomForestClassifier()))
models.append(('DecisionTreeClassifier', DecisionTreeClassifier()))
models.append(('SVM', SVC(kernel='linear')))  #In many SVM we are using linear
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


for name, model in models:
    print(name, model)
    print()
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    print("Confusion matrix: ",confusion_matrix(y_test,y_pred))
    print('\n')
    print("accuracy_score: ",accuracy_score(y_test,y_pred))
    print('\n')
    print("precision_score: ",precision_score(y_test,y_pred))
    print('\n')
    print("recall_score: ",recall_score(y_test,y_pred))
    print('\n')
    print("f1_score: ",f1_score(y_test,y_pred))
    print('\n')
```
